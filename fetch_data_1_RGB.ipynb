```python
import numpy as np
import pandas as pd
import os
import requests
from zipfile import ZipFile
import shutil
from tqdm import tqdm
from datetime import datetime
```

```python
# download data for fall sequences
# http://fenix.ur.edu.pl/~mkepski/ds/data/fall-01-cam0-rgb.zip
url_prefix_fall = "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-"
url_suffix_fall = "-cam0-rgb.zip"
directory = "data/fall_rgb/"
local_filename = os.path.join(directory, "fall-01-cam0-rgb.zip")

# Create the directory if it doesn't exist
os.makedirs(directory, exist_ok=True)

for i in tqdm(range(1,31)):
    start = datetime.now()
    if i<10:
        url = url_prefix_fall + '0' + str(i) + url_suffix_fall
        local_filename = os.path.join(directory, "fall-" + '0' + str(i) + "-cam0-rgb.zip")
    else:
        url = url_prefix_fall + str(i) + url_suffix_fall
        local_filename = os.path.join(directory, "fall-" + str(i) + "-cam0-rgb.zip")
    print(f"Preparing download for url: '{url}'...")
    response = requests.get(url)
    
    with open(local_filename, 'wb') as file:
        file.write(response.content)
    print(f"File '{local_filename}' downloaded successfully.")
    end = datetime.now()
    print(f"Total time taken is: {end-start} seconds\n")
    
#     url = url.replace("cam0", "cam1")
#     local_filename = local_filename.replace("cam0", "cam1")
#     print(f"Preparing download for url: '{url}'...")
#     response = requests.get(url)
    
#     with open(local_filename, 'wb') as file:
#         file.write(response.content)
#     print(f"File '{local_filename}' downloaded successfully.")
```

```python
# unzip all the files to get the frames - fall sequences
# file name example: fall-01-cam0-d.zip
for i in range(1, 31):
    num = str(i)
    if i < 10:
        num = "0" + num
    file_name = "data/fall_rgb/fall-" + num + "-cam" + str(0) + "-rgb.zip"
    with ZipFile(file_name, 'r') as zip: 
        zip.printdir() 
        print('Extracting all the files now...') 
        zip.extractall('data/fall_rgb/') 
        print('Done!')
```

```python
# remove all the .zip folders

def delete_folders_with_extension(directory, extension):
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if item_path.endswith(extension):
            os.remove(item_path)

delete_folders_with_extension("./data/fall_rgb/", ".zip")
delete_folders_with_extension("./data/adl_rgb/", ".zip")
print(f"All files with .zip extension removed.")
```

```python
# remove the folder structure and put all the files in one folder for both fall and adl images

def flatten_directory(src_directory, dest_directory):
    # Create the destination directory if it doesn't exist
    os.makedirs(dest_directory, exist_ok=True)

    # Walk through the source directory
    for root, _, files in os.walk(src_directory):
        for file in files:
            # Create the full source path
            src_path = os.path.join(root, file)

            # Create the full destination path
            dest_path = os.path.join(dest_directory, file)

            # copy the file to the destination directory
            shutil.copy(src_path, dest_path)
            
            os.remove(src_path)
    shutil.rmtree(src_directory)
            
flatten_directory("./data/fall_rgb/", "./data/fall_imgs_rgb")
flatten_directory("./data/adl_rgb/", "./data/adl_imgs_rgb")
```

```python
from PIL import Image
import os

# Specify the input directory
input_directory = './data/adl_imgs_rgb/'

# Specify the target size
target_size = (64, 64)

# Loop through all files in the input directory
for filename in tqdm(os.listdir(input_directory)):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Construct the full file paths
        input_path = os.path.join(input_directory, filename)

        # Open the image
        image = Image.open(input_path)

        # Convert the image to grayscale
        grayscale_image = image.convert('L')

        # Resize the image
        resized_image = grayscale_image.resize(target_size, Image.BICUBIC)

        # Save the resized and grayscale image, overwriting the original
        resized_image.save(input_path)

print("Converting to grayscale, resizing, and replacing complete.")
```
