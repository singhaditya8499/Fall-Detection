```python
import numpy as np
import pandas as pd
import os
import requests
from zipfile import ZipFile
import shutil
```

```python
# download data for fall sequences

url_prefix_fall = "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-"
url_suffix_fall = "-cam0-d.zip"
directory = "data/fall/"
local_filename = os.path.join(directory, "fall-01-cam0-d.zip")

# Create the directory if it doesn't exist
os.makedirs(directory, exist_ok=True)

for i in range(1,31):
    if i<10:
        url = url_prefix_fall + '0' + str(i) + url_suffix_fall
        local_filename = os.path.join(directory, "fall-" + '0' + str(i) + "-cam0-d.zip")
    else:
        url = url_prefix_fall + str(i) + url_suffix_fall
        local_filename = os.path.join(directory, "fall-" + str(i) + "-cam0-d.zip")
    print(f"Preparing download for url: '{url}'...")
    response = requests.get(url)
    
    with open(local_filename, 'wb') as file:
        file.write(response.content)
    print(f"File '{local_filename}' downloaded successfully.")
    
    url = url.replace("cam0", "cam1")
    local_filename = local_filename.replace("cam0", "cam1")
    print(f"Preparing download for url: '{url}'...")
    response = requests.get(url)
    
    with open(local_filename, 'wb') as file:
        file.write(response.content)
    print(f"File '{local_filename}' downloaded successfully.")
```

```python
# download data for active daily life sequences

url_prefix_adl = "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-"
url_suffix_adl = "-cam0-d.zip"
directory = "data/adl/"

# Create the directory if it doesn't exist
os.makedirs(directory, exist_ok=True)

for i in range(1,41):
    if i<10:
        url = url_prefix_adl + '0' + str(i) + url_suffix_adl
        local_filename = os.path.join(directory, "adl-" + '0' + str(i) + "-cam0-d.zip")
    else:
        url = url_prefix_adl + str(i) + url_suffix_adl
        local_filename = os.path.join(directory, "adl-" + str(i) + "-cam0-d.zip")
    print(f"Preparing download for url: '{url}'...")
    response = requests.get(url)
    
    with open(local_filename, 'wb') as file:
        file.write(response.content)
    print(f"File '{local_filename}' downloaded successfully.")
```

```python
# unzip all the files to get the frames - fall sequences
# file name example: fall-01-cam0-d.zip
for i in range(1, 31):
    num = str(i)
    if i < 10:
        num = "0" + num
    file_name = "data/fall/fall-" + num + "-cam" + str(0) + "-d.zip"
    with ZipFile(file_name, 'r') as zip: 
        zip.printdir() 
        print('Extracting all the files now...') 
        zip.extractall('data/fall/') 
        print('Done!')
        
    file_name = "data/fall/fall-" + num + "-cam" + str(1) + "-d.zip"
    with ZipFile(file_name, 'r') as zip: 
        zip.printdir() 
        print('Extracting all the files now...') 
        zip.extractall('data/fall/') 
        print('Done!')
```

```python
# unzip all the files to get the frames - fall sequences
# file name example: adl-01-cam0-d.zip
for i in range(1, 41):
    num = str(i)
    if i < 10:
        num = "0" + num
    file_name = "data/adl/adl-" + num + "-cam" + str(0) + "-d.zip"
    with ZipFile(file_name, 'r') as zip: 
        zip.printdir() 
        print('Extracting all the files now...') 
        zip.extractall('data/adl/') 
        print('Done!')
```

```python
# remove all the .zip folders

def delete_folders_with_extension(directory, extension):
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if item_path.endswith(extension):
            os.remove(item_path)

delete_folders_with_extension("./data/fall/", ".zip")
delete_folders_with_extension("./data/adl/", ".zip")
print(f"All files with .zip extension removed.")
```
