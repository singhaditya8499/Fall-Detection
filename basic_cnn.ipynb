```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from random import randint
```

```python
data = pd.read_csv("./data/labels.csv")
```

```python
data_shape = data.shape
```

```python
data_shape[0]
```

```python
index = [(randint(0, data_shape[0])) for i in range(10)]
```

```python
index
```

```python
fig = plt.figure(figsize=(20,8))
for i, ind in enumerate(index):
    fig.add_subplot(2, 5, i+1)
    image = cv2.imread(data.iloc[ind].image_sample)
    plt.imshow(image)
    plt.axis('off')
    plt.title(data.iloc[ind].label)
```
